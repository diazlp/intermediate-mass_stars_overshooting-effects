```python
#Diagram Kelimpahan Unsur bintang bermassa sedang untuk bintang OV dan bintang NOOV (dari protobintang sampai akhir evolusinya)

#list unsur-unsur penyusun inti
unsur_inti = ['center_h1', 'center_he4', 'center_c12', 'center_o16', 'center_ne20']

#list untuk warna plot
list_warna = ['blue', 'red', 'lime', 'indigo', 'black']

#label untuk penamaan inline-label pada bintang NOOV
label_unsur = ['H', 'He', 'C', 'O', 'Ne']

#fungsi untuk membuat diagram kelimpahan unsur inti bintang dari protobintang sampai akhir evolusinya (default massa bintang = 2Msun)
def core_abundances(var, var_OV, num=2):
    for i in range(len(list_warna)):
        plt.plot(var_OV.data('star_age'), var_OV.data(unsur_inti[i]), c=list_warna[i], linestyle='dashed')
        plt.plot(var.data('star_age'), var.data(unsur_inti[i]), c=list_warna[i], label=label_unsur[i])
    plt.title(f'Diagram Kelimpahan Unsur Bintang {num}$M_\odot$')
    plt.xlabel('Umur Bintang')     
    plt.ylabel('Kelimpahan Unsur')
    labelLines(plt.gca().get_lines(), zorder=7)

#memanggil fungsi pembuatan diagram kelimpahan unsur inti bintang( dengan parameter: variable tanpa OV, variable dengan OV, massa bintang)
core_abundances(h_8m, h_8m_OV, 8)
```
