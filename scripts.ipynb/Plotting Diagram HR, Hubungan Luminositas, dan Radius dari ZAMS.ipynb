```python
#modul yang digunakan
import mesa_reader as ms
import matplotlib.pyplot as plt


#mencari model number pada saat fase ZAMS (deret utama dimulai)
def find_ZAMS_age(var):
    for i in range(1,len(var.star_age)):
         if var.data('center_h1')[i] < 0.6:
            return int(var.data("model_number")[i])
```

```python
#variabel untuk bintang bermassa sedang tanpa OV
h_2m = ms.MesaData('../LOGS/NOOV STARS/2Msun(NOOV)/trimmed_history.data')
h_3m = ms.MesaData('../LOGS/NOOV STARS/3Msun(NOOV)/trimmed_history.data')
h_4m = ms.MesaData('../LOGS/NOOV STARS/4Msun(NOOV)/trimmed_history.data')
h_5m = ms.MesaData('../LOGS/NOOV STARS/5Msun(NOOV)/trimmed_history.data')
h_6m = ms.MesaData('../LOGS/NOOV STARS/6Msun(NOOV)/trimmed_history.data')
h_7m = ms.MesaData('../LOGS/NOOV STARS/7Msun(NOOV)/trimmed_history.data')
h_8m = ms.MesaData('../LOGS/NOOV STARS/8Msun(NOOV)/trimmed_history.data')


#variabel untuk bintang bermassa sedang dengan OV
h_2m_OV = ms.MesaData('../LOGS/OV STARS/2Msun(OV)/trimmed_history.data')
h_3m_OV = ms.MesaData('../LOGS/OV STARS/3Msun(OV)/trimmed_history.data')
h_4m_OV = ms.MesaData('../LOGS/OV STARS/4Msun(OV)/trimmed_history.data')
h_5m_OV = ms.MesaData('../LOGS/OV STARS/5Msun(OV)/trimmed_history.data')
h_6m_OV = ms.MesaData('../LOGS/OV STARS/6Msun(OV)/trimmed_history.data')
h_7m_OV = ms.MesaData('../LOGS/OV STARS/7Msun(OV)/trimmed_history.data')
h_8m_OV = ms.MesaData('../LOGS/OV STARS/8Msun(OV)/trimmed_history.data')
```

```python
#Hubungan luminositas terhadap waktu untuk beragam massa bintang bermassa sedang (dari ZAMS sampai akhir evolusinya)

#list untuk menampung data mulai dari ZAMS sampai akhir evolusi bintang OV dan bintang NOOV
log_L_ZAMS, star_age_ZAMS, log_L_ZAMS_OV, star_age_ZAMS_OV = [], [], [], []


#fungsi untuk plotting hubungan luminositas dengan modul matplotlib
def luminosity_plotter(num):
    plt.plot(star_age_ZAMS_OV, log_L_ZAMS_OV, c='blue', label=f'M={num}$M_\odot$$_,$$_O$$_V$')
    plt.plot(star_age_ZAMS, log_L_ZAMS, c='red',label=f'M={num}$M_\odot$$_,$$_N$$_O$$_O$$_V$')
    plt.title(f'Perubahan Luminositas terhadap Waktu untuk Bintang {num}$M_\odot$')
    plt.xlabel('Umur Bintang')
    plt.ylabel(r'$\log\,L/L_\odot$')
    plt.legend()
    plt.show()


#fungsi untuk membuat hubungan luminositas terhadap waktu dari fase ZAMS hingga akhir evolusi (default massa bintang = 2Msun)
def create_luminosity_from_ZAMS(var, var_OV, num=2):
    for i in range(find_ZAMS_age(var),len(var.star_age)):
            log_L_ZAMS.append(var.data_at_model_number('log_L',i))
            star_age_ZAMS.append(var.data_at_model_number('star_age',i))
    for i in range(find_ZAMS_age(var_OV),len(var_OV.star_age)):
            log_L_ZAMS_OV.append(var_OV.data_at_model_number('log_L',i))
            star_age_ZAMS_OV.append(var_OV.data_at_model_number('star_age',i))
    luminosity_plotter(num)


#memanggil fungsi pembuatan hubungan luminositas dari fase ZAMS( dengan parameter: variable tanpa OV, variable dengan OV, massa bintang)
create_luminosity_from_ZAMS(h_8m,h_8m_OV,8)
```

```python
#Hubungan radius terhadap waktu untuk beragam massa bintang bermassa sedang (dari ZAMS sampai akhir evolusinya)

#list untuk menampung data mulai dari ZAMS sampai akhir evolusi bintang OV dan bintang NOOV
log_R_ZAMS, star_age_ZAMS, log_R_ZAMS_OV, star_age_ZAMS_OV = [], [], [], []


        
#fungsi untuk plotting hubungan luminositas dengan modul matplotlib
def radius_plotter(num):
    plt.plot(star_age_ZAMS_OV, log_R_ZAMS_OV, c='blue', label=f'M={num}$M_\odot$$_,$$_O$$_V$')
    plt.plot(star_age_ZAMS, log_R_ZAMS, c='red',label=f'M={num}$M_\odot$$_,$$_N$$_O$$_O$$_V$')
    plt.title(f'Perubahan Radius terhadap Waktu untuk Bintang {num}$M_\odot$')
    plt.xlabel('Umur Bintang')
    plt.ylabel(r'Radius ($R_\odot$)')
    plt.legend()
    plt.show()

#fungsi untuk membuat hubungan radius terhadap waktu dari fase ZAMS hingga akhir evolusi (default massa bintang = 2Msun)
def create_radius_from_ZAMS(var, var_OV, num=2):
    for i in range(find_ZAMS_age(var),len(var.star_age)):
            log_R_ZAMS.append(var.data_at_model_number('log_R',i))
            star_age_ZAMS.append(var.data_at_model_number('star_age',i))
    for i in range(find_ZAMS_age(var_OV),len(var_OV.star_age)):
            log_R_ZAMS_OV.append(var_OV.data_at_model_number('log_R',i))
            star_age_ZAMS_OV.append(var_OV.data_at_model_number('star_age',i))
    radius_plotter(num)


#memanggil fungsi pembuatan hubungan radius dari fase ZAMS( dengan parameter: variable tanpa OV, variable dengan OV, massa bintang)
create_radius_from_ZAMS(h_8m,h_8m_OV,8)
```

```python
#Diagram HR untuk beragam massa bintang bermassa sedang(dari ZAMS sampai akhir evolusinya)

#list untuk menampung data mulai dari ZAMS sampai akhir evolusi bintang OV dan bintang NOOV
log_Teff_ZAMS, log_L_ZAMS, log_Teff_ZAMS_OV, log_L_ZAMS_OV = [], [], [], []

        

#fungsi untuk plotting dengan modul matplotlib
def HR_diagram_plotter(num):
    plt.plot(log_Teff_ZAMS_OV, log_L_ZAMS_OV, c='blue', label=f'M={num}$M_\odot$$_,$$_O$$_V$')
    plt.plot(log_Teff_ZAMS, log_L_ZAMS, c='red',label=f'M={num}$M_\odot$$_,$$_N$$_O$$_O$$_V$')
    plt.title(f'Diagram HR Bintang Bermassa Sedang {num}$M_\odot$')
    plt.xlabel(r'$\log T_{\mathrm{\mathrm{eff}}}$')
    plt.ylabel(r'$\log\,L/L_\odot$')
    plt.gca().invert_xaxis()
    plt.legend()
    plt.show()


#fungsi untuk membuat diagram HR dari fase ZAMS hingga akhir evolusi (default massa bintang = 2Msun)
def create_HR_from_ZAMS(var, var_OV, num=2):
    for i in range(find_ZAMS_age(var),len(var.star_age)):
            log_Teff_ZAMS.append(var.data_at_model_number('log_Teff',i))
            log_L_ZAMS.append(var.data_at_model_number('log_L',i))
    for i in range(find_ZAMS_age(var_OV), len(var_OV.star_age)):
            log_Teff_ZAMS_OV.append(var_OV.data_at_model_number('log_Teff',i))
            log_L_ZAMS_OV.append(var_OV.data_at_model_number('log_L',i))
    HR_diagram_plotter(num)


#memanggil fungsi pembuatan diagram HR dari fase ZAMS( dengan parameter: variable tanpa OV, variable dengan OV, massa bintang)
create_HR_from_ZAMS(h_8m,h_8m_OV,8)
```
