```python
#modul-modul yang dibutuhkan
import mesa_reader as ms
import matplotlib.pyplot as plt
```

```python
#variabel untuk bintang bermassa sedang tanpa OV
h_2m = ms.MesaData('../LOGS/NOOV STARS/2Msun(NOOV)/trimmed_history.data')
h_3m = ms.MesaData('../LOGS/NOOV STARS/3Msun(NOOV)/trimmed_history.data')
h_4m = ms.MesaData('../LOGS/NOOV STARS/4Msun(NOOV)/trimmed_history.data')
h_5m = ms.MesaData('../LOGS/NOOV STARS/5Msun(NOOV)/trimmed_history.data')
h_6m = ms.MesaData('../LOGS/NOOV STARS/6Msun(NOOV)/trimmed_history.data')
h_7m = ms.MesaData('../LOGS/NOOV STARS/7Msun(NOOV)/trimmed_history.data')
h_8m = ms.MesaData('../LOGS/NOOV STARS/8Msun(NOOV)/trimmed_history.data')


#variabel untuk bintang bermassa sedang dengan OV
h_2m_OV = ms.MesaData('../LOGS/OV STARS/2Msun(OV)/trimmed_history.data')
h_3m_OV = ms.MesaData('../LOGS/OV STARS/3Msun(OV)/trimmed_history.data')
h_4m_OV = ms.MesaData('../LOGS/OV STARS/4Msun(OV)/trimmed_history.data')
h_5m_OV = ms.MesaData('../LOGS/OV STARS/5Msun(OV)/trimmed_history.data')
h_6m_OV = ms.MesaData('../LOGS/OV STARS/6Msun(OV)/trimmed_history.data')
h_7m_OV = ms.MesaData('../LOGS/OV STARS/7Msun(OV)/trimmed_history.data')
h_8m_OV = ms.MesaData('../LOGS/OV STARS/8Msun(OV)/trimmed_history.data')

#list untuk data bintang NOOV
data_bintang = [
    h_2m, h_3m, h_4m, h_5m, h_6m, h_7m, h_8m
]

#list untuk data bintang OV
data_bintang_OV = [
    h_2m_OV, h_3m_OV, h_4m_OV, h_5m_OV, h_6m_OV, h_7m_OV, h_8m_OV
]

#list untuk warna plot
list_warna = [
    'blue', 'red', 'lime', 'orange', 'black', 'indigo', 'teal'
]

#list untuk menampung data mulai dari ZAMS sampai fase akhir RGB bintang NO OV
log_Teff2M, log_Teff3M, log_Teff4M, log_Teff5M, log_Teff6M, log_Teff7M, log_Teff8M = [], [], [], [], [], [], []
log_L2M, log_L3M, log_L4M, log_L5M, log_L6M, log_L7M, log_L8M  = [], [], [], [], [], [], []

log_Teff_ZAMS_to_AGB_list = [log_Teff2M, log_Teff3M, log_Teff4M, log_Teff5M, log_Teff6M, log_Teff7M, log_Teff8M]
log_L_ZAMS_to_AGB_list = [log_L2M, log_L3M, log_L4M, log_L5M, log_L6M, log_L7M, log_L8M]

#list untuk menampung data mulai dari ZAMS sampai fase akhir RGB bintang OV
log_Teff2M_OV, log_Teff3M_OV, log_Teff4M_OV, log_Teff5M_OV, log_Teff6M_OV, log_Teff7M_OV, log_Teff8M_OV = [], [], [], [], [], [], []
log_L2M_OV, log_L3M_OV, log_L4M_OV, log_L5M_OV, log_L6M_OV, log_L7M_OV, log_L8M_OV  = [], [], [], [], [], [], []

log_Teff_ZAMS_to_AGB_OV_list = [log_Teff2M_OV, log_Teff3M_OV, log_Teff4M_OV, log_Teff5M_OV, log_Teff6M_OV, log_Teff7M_OV, log_Teff8M_OV]
log_L_ZAMS_to_AGB_OV_list = [log_L2M_OV, log_L3M_OV, log_L4M_OV, log_L5M_OV, log_L6M_OV, log_L7M_OV, log_L8M_OV]
```

```python
#Diagram HR bintang bermassa sedang tanpa efek overshooting

#fungsi untuk pembuatan diagram HR NOOV
def HR_diagram_NOOV(num=2):
    for i in range(len(data_bintang)):
        plt.plot(data_bintang[i].log_Teff,data_bintang[i].log_L, c=list_warna[i], label=f'M={num+i}$M_\odot$$_,$$_N$$_O$$_O$$_V$')
        plt.title(r'Diagram HR Bintang Bermassa Sedang tanpa Efek $\it{Overshooting}$')
        plt.gca().invert_xaxis()
        plt.xlabel(r'$\log T_{\mathrm{\mathrm{eff}}}$')     
        plt.ylabel(r'$\log\,L/L_\odot$')
        plt.legend(loc='upper right', bbox_to_anchor=(1.36, 1))
        plt.show
        
HR_diagram_NOOV()
```

```python
#Diagram HR bintang bermassa sedang dengan efek overshooting

#fungsi untuk pembuatan diagram HR OV
def HR_diagram_OV(num=2):
    for i in range(len(data_bintang_OV)):
        plt.plot(data_bintang_OV[i].log_Teff,data_bintang_OV[i].log_L, c=list_warna[i], label=f'M={num+i}$M_\odot$$_,$$_O$$_V$')
        plt.title(r'Diagram HR Bintang Bermassa Sedang dengan Efek $\it{Overshooting}$')
        plt.gca().invert_xaxis()
        plt.xlabel(r'$\log T_{\mathrm{\mathrm{eff}}}$')     
        plt.ylabel(r'$\log\,L/L_\odot$')
        plt.legend(loc='upper right', bbox_to_anchor=(1.36, 1))
        plt.show
        
HR_diagram_OV()
```

```python
#Diagram HR bintang bermassa sedang untuk bintang OV dan bintang NOOV (dari ZAMS sampai fase awal RGB)

#list untuk menampung data mulai dari ZAMS sampai fase awal AGB bintang NO OV
log_Teff2M, log_Teff3M, log_Teff4M, log_Teff5M, log_Teff6M, log_Teff7M, log_Teff8M = [], [], [], [], [], [], []
log_L2M, log_L3M, log_L4M, log_L5M, log_L6M, log_L7M, log_L8M  = [], [], [], [], [], [], []

log_Teff_ZAMS_to_AGB_list = [log_Teff2M, log_Teff3M, log_Teff4M, log_Teff5M, log_Teff6M, log_Teff7M, log_Teff8M]
log_L_ZAMS_to_AGB_list = [log_L2M, log_L3M, log_L4M, log_L5M, log_L6M, log_L7M, log_L8M]

#list untuk menampung data mulai dari ZAMS sampai fase awal AGB bintang OV
log_Teff2M_OV, log_Teff3M_OV, log_Teff4M_OV, log_Teff5M_OV, log_Teff6M_OV, log_Teff7M_OV, log_Teff8M_OV = [], [], [], [], [], [], []
log_L2M_OV, log_L3M_OV, log_L4M_OV, log_L5M_OV, log_L6M_OV, log_L7M_OV, log_L8M_OV  = [], [], [], [], [], [], []

log_Teff_ZAMS_to_AGB_OV_list = [log_Teff2M_OV, log_Teff3M_OV, log_Teff4M_OV, log_Teff5M_OV, log_Teff6M_OV, log_Teff7M_OV, log_Teff8M_OV]
log_L_ZAMS_to_AGB_OV_list = [log_L2M_OV, log_L3M_OV, log_L4M_OV, log_L5M_OV, log_L6M_OV, log_L7M_OV, log_L8M_OV]


#fungsi untuk mencari model number pada saat fase ZAMS (deret utama dimulai)
def find_ZAMS_age(var):
    for i in range(1,len(var.star_age)):
         if var.data('center_h1')[i] < 0.6:
            return var.data("model_number")[i]
    
#fungsi untuk membuat plot dengan modul matplotlib
def HR_diagram_plotter(var, count=0):
    while count<len(var):
        plt.plot(log_Teff_ZAMS_to_AGB_OV_list[count], log_L_ZAMS_to_AGB_OV_list[count], c='black', linestyle='dashed')
        plt.plot(log_Teff_ZAMS_to_AGB_list[count], log_L_ZAMS_to_AGB_list[count], c='black')
        plt.title(r'Diagram HR Bintang dengan efek $\it{Overshooting}$ dan tanpa efek $\it{Overshooting}$')
        plt.xlabel(r'$\log T_{\mathrm{\mathrm{eff}}}$')
        plt.ylabel(r'$\log\,L/L_\odot$')
        plt.gca().invert_xaxis()
        plt.grid(False)
        count+=1

    
def HR_diagram_ZAMS_to_AGB(var, var_OV):
    var, var_OV = [var[i] for i in range(len(var))], [var_OV[i] for i in range(len(var_OV))]
    count =0
    while count<len(var):
        for i in range(find_ZAMS_age(var[count]), 790):
            log_Teff_ZAMS_to_AGB_list[count].append(var[count].data_at_model_number('log_Teff',i))
            log_L_ZAMS_to_AGB_list[count].append([var[count].data_at_model_number('log_L',i)])        
        
        for i in range(find_ZAMS_age(var_OV[count]), 690):
            log_Teff_ZAMS_to_AGB_OV_list[count].append(var_OV[count].data_at_model_number('log_Teff',i))
            log_L_ZAMS_to_AGB_OV_list[count].append(var_OV[count].data_at_model_number('log_L',i))
            
        count+=1
    HR_diagram_plotter(var)

plt.text(3.99,1.34,r'2M$_\odot$')
plt.text(4.1,2.1,r'3M$_\odot$')
plt.text(4.21,2.5,r'4M$_\odot$')
plt.text(4.28,2.76,r'5M$_\odot$')
plt.text(4.33,3.1,r'6M$_\odot$')
plt.text(4.37,3.3,r'7M$_\odot$')
plt.text(4.32,4,r'8M$_\odot$')
            
HR_diagram_ZAMS_to_AGB(data_bintang, data_bintang_OV)                    
```
