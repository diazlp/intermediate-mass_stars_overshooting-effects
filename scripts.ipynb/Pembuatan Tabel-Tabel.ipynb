```python
#deklarasi fungsi-fungsi yang digunakan

#fungsi untuk menentukan umur pada fase ZAMS bintang (dimulai pada awal pembakaran hidrogen)
def find_ZAMS_age(var):
    for i in range(1,len(var.star_age)):
         if var.data('center_h1')[i] == 0.7:
            return var.data("star_age")[i]

#fungsi untuk menentukan umur pada awal fase RGB bintang (dimulai saat penghabisan inti hidrogen)
def find_RGB_age(var):
    for i in range(1,len(var.star_age)):
         if var.data('center_h1')[i] == 0:
            return var.data("star_age")[i]

        
#fungsi untuk menentukan umur pada fase awal AGB bintang (dimulai saat penghabisan inti helium)
def find_AGB_age(var):
    for i in range(1,len(var.star_age)):
         if var.data('center_he4')[i] <= 1e-4:
            return var.data("star_age")[i]


#fungsi untuk menentukan masa hidup bintang pada fase DU (dari ZAMS hingga awal RGB)
def MS_lifetime(var):
    return find_RGB_age(var)-find_ZAMS_age(var)



##########################################################################################



#fungsi untuk menentukan kelimpahan inti karbon pada akhir masa hidup bintang 
def carbon_end(var):
    return(float(var.data('center_c12')[-1]))


#fungsi untuk menentukan kelimpahan inti oksigen pada akhir masa hidup bintang 
def oxygen_end(var):
    return(float(var.data('center_o16')[-1]))

#fungsi untuk menentukan kelimpahan inti neon pada akhir masa hidup bintang 
def neon_end(var):
    return(float(var.data('center_ne20')[-1]))



##########################################################################################



####################################### TEST RUN #########################################


# print(neon_end(h_2m)/carbon_end(h_2m))


# for i in range(1,len(h_2m.star_age)):
#     print(h_2m.data('center_c12')[-1])


# print(MS_lifetime(h_2m))

# print(find_RGB_age(h_2m)
# print(find_RGB_age(h_2m_OV))

# print(find_ZAMS_age(h_2m))

# print(find_AGB_age(h_2m))
# print(find_AGB_age(h_2m_OV))
```

```python
#pembuatan tabel umur pada saat fase X untuk bintang NOOV

#list untuk massa bintang
massa_bintang = [
    h_2m_mass, h_3m_mass, h_4m_mass, h_5m_mass, h_6m_mass, h_7m_mass, h_8m_mass
]

#list untuk data bintang
data_bintang = [
    h_2m, h_3m, h_4m, h_5m, h_6m, h_7m, h_8m
]



#masukkan tabel
masukan_tabel_umur_faseX_NOOV= {
    r'Massa Bintang ($M_{\odot,NOOV}$)': [massa_bintang[i] for i in range(len(massa_bintang))],
    'ZAMS (tahun)': ["%.2e" % find_ZAMS_age(data_bintang[i]) for i in range(len(massa_bintang))],
    'RGB (tahun)': ["%.2e" % find_RGB_age(data_bintang[i]) for i in range(len(massa_bintang))],
    'AGB (tahun)': ["%.2e" % find_AGB_age(data_bintang[i]) for i in range(len(massa_bintang))],
    'Masa Hidup Fase DU (tahun)': ["%.2e" % MS_lifetime(data_bintang[i]) for i in range(len(massa_bintang))],   
}


tabel_umur_faseX_NOOV = pd.DataFrame(masukan_tabel_umur_faseX_NOOV)
display(tabel_umur_faseX_NOOV)
```

```python
#pembuatan tabel umur bintang pada akhir masa hidupnya untuk bintang OV dan bintang NOOV

#list untuk menampung data umur bintang pada akhir masa hidupnya
umur_bintang = []
umur_bintang_OV = []

for i in range(len(data_bintang)):
    umur_bintang.append(data_bintang[i].data('star_age')[-1])
    umur_bintang_OV.append(data_bintang_OV[i].data('star_age')[-1])
    
#masukkan tabel
masukan_tabel_umur_akhir_bintang = {
    r'Massa Bintang ($M_{\odot}$)' : [massa_bintang[i] for i in range(len(massa_bintang))]  + [massa_bintang_OV[i] for i in range(len(massa_bintang))],
    '' : [r'$NO OV$']*7 + [r'$OV$']*7,
    'Umur Bintang (tahun)' : ['%.2e' %umur_bintang[i] for i in range(len(umur_bintang))] + ['%.2e' % umur_bintang_OV[i] for i in range(len(umur_bintang))],
}
tabel_umur_akhir_bintang = pd.DataFrame(masukan_tabel_umur_akhir_bintang)
tabel_umur_akhir_bintang_CLEANED =  pd.pivot(tabel_umur_akhir_bintang, index=r'Massa Bintang ($M_{\odot}$)', columns='', values=['Umur Bintang (tahun)'])

display(tabel_umur_akhir_bintang_CLEANED)
```
